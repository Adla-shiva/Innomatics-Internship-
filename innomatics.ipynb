```python
import requests
from bs4 import BeautifulSoup

def get_imdb_id(movie_title):
    # Replace spaces in the movie title with '+' for the IMDb search URL
    search_query = '+'.join(movie_title.split())

    # IMDb search URL
    imdb_search_url = f'https://www.imdb.com/find?q={search_query}&s=tt'

    # Send a GET request to IMDb
    response = requests.get(imdb_search_url)

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the first result (assuming it's the most relevant)
    result = soup.find('td', class_='result_text')

    # Extract the IMDb ID from the result
    if result:
        imdb_id = result.find('a')['href'].split('/')[2]
        return imdb_id
    else:
        return None

# Example usage
movie_title = "The Godfather"
imdb_id = get_imdb_id(movie_title)

if imdb_id:
    print(f"The IMDb ID for '{movie_title}' is {imdb_id}")
else:
    print(f"No results found for '{movie_title}'")
```
